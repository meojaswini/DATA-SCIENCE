# Get How Many Times a Item is Bought at respective UnitPrice in Descending Order

```python
total_purchase=df[['StockCode','Quantity']]
total_purchase=total_purchase.groupby('StockCode')
total_purchase=total_purchase.sum()
total_purchase=total_purchase.sort_values(by='Quantity',ascending=False)
total_purchase=total_purchase[total_purchase.Quantity > total_purchase['Quantity'].median()]
total_purchase=total_purchase.reset_index()
most=total_purchase[:]
least=total_purchase[:]
final=pd.merge(most,least,on='StockCode',how='outer')
final=final['StockCode'].to_frame()
new=pd.merge(final,df,on='StockCode',how='inner')
new=new[['StockCode','UnitPrice','Quantity']]
new=new.groupby('StockCode')['UnitPrice'].value_counts().to_frame()
new=new.rename(columns={'UnitPrice':'NooftimesBought'})
new=new.reset_index().groupby('StockCode').sum()
new=new.sort_values(by='NooftimesBought',ascending=False)
```

# Merge the Stockcode and Retrived data from original dataset & Groupby into Invoice No and Description based on Quantity Bought and map into a Matrix

# Use apriori() method to get itemsets and respective support

```python
name=apriori(test,min_support=0.01,use_colnames=True,low_memory=True) #low memory is used to avoid any memory error
name
```

# Use association_rules() to get rules showing antecedant and consequent 

```python
rules = association_rules(name)
rules
```

import warnings
warnings.filterwarnings("ignore", 'This pattern has match groups')

# Convert frozen set type antecedants and consequents to string

```python
rules["antecedents"] = rules["antecedents"].apply(lambda x: ', '.join(list(x))).astype("unicode")
rules["consequents"] = rules["consequents"].apply(lambda x: ', '.join(list(x))).astype("unicode")
```

# Design a simple if-else Recommendation system to predict the items bought along with the inputed item

```python
def Recomend(items):
        df1=rules[reduce(lambda x,y:x&y, (rules['antecedents'].str.contains(s) for s in items))]
        dict1=list(set(df1['consequents'].to_list()))
        return dict1
```
